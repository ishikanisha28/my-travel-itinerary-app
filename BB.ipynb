```python
import os
import streamlit as st
import openai
from fpdf import FPDF
import unicodedata

# ‚úÖ Fetch OpenAI API Key from environment variable
api_key = os.getenv("OPENAI_API_KEY")

if not api_key:
    st.error("‚ö†Ô∏è ERROR: OpenAI API Key is missing! Set it using environment variables.")
    st.stop()

# ‚úÖ Set OpenAI API Key properly
openai.api_key = api_key

# ‚úÖ Set page configuration
st.set_page_config(page_title="Travel Itinerary Generator", layout="wide")

# ‚úÖ Sidebar content
st.sidebar.title("‚ÑπÔ∏è About This Program")
st.sidebar.info(
    "**Travel Itinerary Generator**\n\n"
    "üîπ Uses AI to create customized travel plans.\n\n"
    "üîπ Generates detailed itineraries with activity and food suggestions.\n\n"
    "üîπ Download itineraries directly as PDF files.\n\n"
    "_Plan your perfect trip effortlessly!_"
)

# ‚úÖ Function to generate itinerary using OpenAI API with custom options
def generate_itinerary(location, days, month, budget, activities, travel_companion):
    activity_str = ", ".join(activities) if activities else "any"
    prompt = (
        f"Generate a {days}-day travel itinerary for {location} in {month} for a {budget} budget. "
        f"Focus on {activity_str} activities for someone traveling {travel_companion}. "
        "Include morning, afternoon, and evening suggestions with food options. "
        "Ensure popular and offbeat spots are covered with specific timings. "
        "Avoid displaying prices."
    )
    try:
        response = openai.ChatCompletion.create(
            model="gpt-4",
            messages=[
                {"role": "system", "content": "You are a helpful travel assistant."},
                {"role": "user", "content": prompt}
            ]
        )
        return response["choices"][0]["message"]["content"]
    except openai.error.OpenAIError as e:
        return f"‚ö†Ô∏è Unable to generate itinerary. Error: {str(e)}"

# ‚úÖ Function to remove non-ASCII characters
def remove_non_ascii(text):
    return unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8')

# ‚úÖ Function to create PDF
def create_pdf(itinerary, location, days, month):
    pdf = FPDF()
    pdf.add_page()
    pdf.set_font("Arial", size=12)

    # ‚úÖ Add a title to the PDF
    pdf.set_font("Arial", 'B', 16)
    pdf.cell(200, 10, txt=f"{days}-Day Travel Itinerary for {location} in {month}", ln=True, align='C')
    pdf.ln(10)

    # ‚úÖ Ensure UTF-8 encoding and add itinerary content
    itinerary_clean = remove_non_ascii(itinerary)
    pdf.set_font("Arial", size=12)
    pdf.multi_cell(0, 10, itinerary_clean)

    # ‚úÖ Save PDF to bytes
    pdf_bytes = pdf.output(dest='S').encode('latin-1')
    return pdf_bytes

# ‚úÖ Main application logic
def main():
    st.title("üåç Travel Itinerary Generator ‚úàÔ∏è")
    st.subheader("Plan your perfect trip with AI!")

    # Initialize session state for itinerary if not present
    if "itinerary" not in st.session_state:
        st.session_state["itinerary"] = None

    # Input fields for location, days, and month
    location = st.text_input("üìç Enter the location:")
    days = st.number_input("üìÖ Enter the number of days (1-7):", min_value=1, max_value=7, value=2)
    month = st.selectbox(
        "üóìÔ∏è Select the month of your trip:",
        ["January", "February", "March", "April", "May", "June",
         "July", "August", "September", "October", "November", "December"]
    )

    # ‚úÖ Customization options
    budget = st.selectbox("üí∏ Choose your budget level:", ["Budget", "Mid-range", "Luxury"])
    activities = st.multiselect("üéØ Choose activities you like:", ["Adventure", "Relaxation", "Cultural", "Sightseeing", "Food Tour"])
    travel_companion = st.selectbox("üë• Who are you traveling with?", ["Solo", "Couple", "Family", "Friends"])

    # ‚úÖ Generate itinerary when button is clicked
    if st.button("üöÄ Generate Itinerary"):
        if location.strip():
            st.session_state["itinerary"] = None  # ‚úÖ Reset stored itinerary
            itinerary = generate_itinerary(location, days, month, budget, activities, travel_companion)
            st.session_state["itinerary"] = itinerary
            st.success("‚úÖ Here is your itinerary:")
            st.markdown(itinerary)
        else:
            st.warning("‚ö†Ô∏è Please enter a valid location.")

    # ‚úÖ Enable PDF download only if an itinerary has been generated
    if st.session_state["itinerary"]:
        pdf_bytes = create_pdf(st.session_state["itinerary"], location, days, month)
        st.download_button(
            label="üì• Download Itinerary as PDF",
            data=pdf_bytes,
            file_name=f"{location}_Travel_Itinerary.pdf",
            mime="application/pdf"
        )
    else:
        st.info("Generate your itinerary first to enable PDF download.")

# ‚úÖ Run the main function
if __name__ == "__main__":
    main()
```
